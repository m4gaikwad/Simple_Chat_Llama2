## Imports

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, TextStreamer
import torch
```

## Model

Can use "meta-llama/Llama-2-70b-chat-hf" if you have permission or can use below.

```python
name = "NousResearch/Llama-2-7b-chat-hf"
```

## Tokenizer and Model

```python
tokenizer = AutoTokenizer.from_pretrained(name, cache_dir="./model/", use_auth_token=auth_token)
```

```python
model = AutoModelForCausalLM.from_pretrained(name, cache_dir="./model/", use_auth_token=auth_token,
                                             rope_scaling={"type": "dynamic", "factor": 2},
                                             torch_dtype=torch.float16, load_in_8bit= True)
```

## Prompt

```python
prompt = "### User: How random forest algorithm works  \
          ### Assistant:"

# Prompt to Tokenizer
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# TextStreamer to pass inputs to model
streamer = TextStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)
```

## Output

```python
output = model.generate(**inputs, streamer=streamer,
                        use_cache=True, max_new_tokens=float('inf'))

output_text = tokenizer.decode(output[0], skip_special_tokens=True)
```
